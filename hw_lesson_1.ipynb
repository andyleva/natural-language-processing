3. Изменим регистр твитов на нижний с помощью .lower()

```python
def remove_pattern(text,pattern):     
    #r = set(re.findall(pattern,text))
    r = set(pattern.findall(text))    
    if len(r)!=0:
        for i in r:            
            text_old = text                        
            try:
                text_new = re.sub(i," ",str(text_old))                
                if text_new.replace(" ","")=="":
                    text=text_old
                else:
                    text=text_new                
            except:
                text=text_old                
    text.lower()
    return text
```

```python
combine_df['tweet_txt'] = np.vectorize(remove_pattern)(combine_df['tweet'],re.compile("@[\w]*"))
```

4. Заменим сокращения с апострофами (пример: ain't, can't) на пробел, используя apostrophe_dict. Для этого необходимо сделать функцию: для каждого слова в тексте проверить (for word in text.split()), если слово есть в словаре apostrophe_dict в качестве ключа (сокращенного слова), то заменить ключ на значение (полную версию слова).

```python
def multiple_replace(target_str, replace_values):

    for i, j in replace_values.items():
        # меняем все target_str на подставляемое
        target_str = target_str.replace(i, j)
        #print(i,j)
    return target_str

combine_df['tweet_txt'] = np.vectorize(multiple_replace)(combine_df['tweet_txt'],apostrophe_dict)  
```

### 12. Удалим стоп-слова из токенов, используя nltk.corpus.stopwords. Создадим столбец 'tweet_token_filtered' без стоп-слов.

```python
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
def tweet_token_filtered(text):
    stop_words = stopwords.words("english")
    words = tknz.word_tokenize(text)
    without_stop_words = [word for word in words if not word in stop_words]
    return without_stop_words
```

```python
df['tweet_token_filtered']= df['tweet_4'].apply(tweet_token_filtered)
```

13. Применим стемминг к токенам с помощью nltk.stem.PorterStemmer. Создадим столбец 'tweet_stemmed' после применения стемминга.

```python
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.corpus import wordnet
from functools import lru_cache
from multiprocessing import Pool
from tqdm import tqdm_notebook as tqdm
```

```python
nltk.download('wordnet')
```

```python
def tweet_stemmed(text):
    stemmer = PorterStemmer()    
    text_stemmer = " ".join([stemmer.stem(w) for w in text])
    return text_stemmer
```

```python
df['tweet_stemmed']= df['tweet_token_filtered'].apply(tweet_stemmed)
```

```python
df.head(3)
```

14. Применим лемматизацию к токенам с помощью nltk.stem.wordnet.WordNetLemmatizer. Создадим столбец 'tweet_lemmatized' после применения лемматизации.

```python
def tweet_lemmatize(text):
    lemmatizer=WordNetLemmatizer()   
    text_stemmer = " ".join([lemmatizer.lemmatize(w, wordnet.VERB) for w in text])
    return text_stemmer
```

```python
df['tweet_lemmatize']= df['tweet_token_filtered'].apply(tweet_lemmatize)
```

```python
def clean_data(data_df):
    data_df['tweet_txt'] = np.vectorize(remove_pattern)(data_df['tweet'],re.compile("@[\w]*"))
    data_df['tweet_txt'] = np.vectorize(multiple_replace)(data_df['tweet_txt'],short_word_dict)
    data_df['tweet_txt'] = np.vectorize(multiple_replace)(data_df['tweet_txt'],emoticon_dict)
    data_df['tweet_2'] = np.vectorize(remove_pattern)(data_df['tweet_txt'],re.compile("[^\w]"))
    data_df['tweet_3'] = np.vectorize(remove_pattern)(data_df['tweet_2'],re.compile("[^a-zA-Z0-9]"))
    data_df['tweet_4'] = np.vectorize(remove_pattern)(data_df['tweet_3'],re.compile("[^a-zA-Z]"))
    data_df['tweet_no_one'] = data_df['tweet_4'].apply(delete_one_symbol)
    data_df['tweet_token'] = data_df['tweet_no_one'].apply(tweet_token)
    data_df['tweet_token_filtered'] = data_df['tweet_4'].apply(tweet_token_filtered)
    data_df['tweet_stemmed'] = data_df['tweet_token_filtered'].apply(tweet_stemmed)
    data_df['tweet_lemmatize']= data_df['tweet_token_filtered'].apply(tweet_lemmatize)
    return data_df
```
